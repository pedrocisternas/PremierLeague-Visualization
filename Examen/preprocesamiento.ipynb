MATCHES

```python
df_matches = pd.read_csv('england-premier-league-matches-2018-to-2019-stats.csv')
df_matches.head()
```

```python
def label_initial_h (row):
  if row['home_team_name'] == 'Manchester United' :
      return 'MUN'
  elif row['home_team_name'] == 'Manchester City' :
      return 'MCI'
  elif row['home_team_name'] == 'Arsenal' :
      return 'ARS'
  elif row['home_team_name'] == 'Leicester City' :
      return 'LEI'
  elif row['home_team_name'] == 'Crystal Palace' :
      return 'CRY'
  elif row['home_team_name'] == 'Everton' :
      return 'EVE'
  elif row['home_team_name'] == 'Burnley' :
      return 'BUR'
  elif row['home_team_name'] == 'AFC Bournemouth' :
      return 'BOU'
  elif row['home_team_name'] == 'Southampton' :
      return 'SOU'
  elif row['home_team_name'] == 'Tottenham Hotspur' :
      return 'TOT'
  elif row['home_team_name'] == 'Liverpool' :
      return 'LIV'
  elif row['home_team_name'] == 'Chelsea' :
      return 'CHE'
  elif row['home_team_name'] == 'West Ham United' :
      return 'WHU'
  elif row['home_team_name'] == 'Watford' :
      return 'WAT'
  elif row['home_team_name'] == 'Newcastle United' :
      return 'NEW'
  elif row['home_team_name'] == 'Cardiff City' :
      return 'CAR'
  elif row['home_team_name'] == 'Fulham' :
      return 'FUL'
  elif row['home_team_name'] == 'Brighton & Hove Albion' :
      return 'BHA'
  elif row['home_team_name'] == 'Huddersfield Town' :
      return 'HUD'
  elif row['home_team_name'] == 'Wolverhampton Wanderers' :
      return 'WOL'
```

```python
def label_initial_a (row):
  if row['away_team_name'] == 'Manchester United' :
      return 'MUN'
  elif row['away_team_name'] == 'Manchester City' :
      return 'MCI'
  elif row['away_team_name'] == 'Arsenal' :
      return 'ARS'
  elif row['away_team_name'] == 'Leicester City' :
      return 'LEI'
  elif row['away_team_name'] == 'Crystal Palace' :
      return 'CRY'
  elif row['away_team_name'] == 'Everton' :
      return 'EVE'
  elif row['away_team_name'] == 'Burnley' :
      return 'BUR'
  elif row['away_team_name'] == 'AFC Bournemouth' :
      return 'BOU'
  elif row['away_team_name'] == 'Southampton' :
      return 'SOU'
  elif row['away_team_name'] == 'Tottenham Hotspur' :
      return 'TOT'
  elif row['away_team_name'] == 'Liverpool' :
      return 'LIV'
  elif row['away_team_name'] == 'Chelsea' :
      return 'CHE'
  elif row['away_team_name'] == 'West Ham United' :
      return 'WHU'
  elif row['away_team_name'] == 'Watford' :
      return 'WAT'
  elif row['away_team_name'] == 'Newcastle United' :
      return 'NEW'
  elif row['away_team_name'] == 'Cardiff City' :
      return 'CAR'
  elif row['away_team_name'] == 'Fulham' :
      return 'FUL'
  elif row['away_team_name'] == 'Brighton & Hove Albion' :
      return 'BHA'
  elif row['away_team_name'] == 'Huddersfield Town' :
      return 'HUD'
  elif row['away_team_name'] == 'Wolverhampton Wanderers' :
      return 'WOL'
```

```python
df_matches_filtrado = df_matches[['timestamp', 'date_GMT', 'attendance', 'home_team_name', 'away_team_name', 'home_team_goal_count', 'away_team_goal_count', 'referee', 'Game Week', 'home_team_goal_timings', 'away_team_goal_timings', 'home_team_shots', 'away_team_shots', 'home_team_shots_on_target', 'away_team_shots_on_target', 'team_a_xg', 'team_b_xg', 'stadium_name']]
df_matches_filtrado.head()
```

```python
df_matches_filtrado.rename(columns = {'Game Week':'gameweek'}, inplace = True)
df_matches_filtrado.head()
```

```python
df_matches_filtrado['home_team_initial'] = df_matches.apply (lambda row: label_initial_h(row), axis=1)
```

```python
df_matches_filtrado['away_team_initial'] = df_matches.apply (lambda row: label_initial_a(row), axis=1)
```

```python
df_matches_filtrado.head()
```

```python
df_matches_filtrado.dtypes
```

```python
df_matches_filtrado.to_csv('premier_league_18-19_matches.csv', index=False, header=True)
```
